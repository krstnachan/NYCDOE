    - Assume that in the first year of opening, 2021-2022, the school can hold up to 500 students. The new school will absorb excess students from the most overcrowded school first and then the second most overcrowded school, 25Q032 and 25Q184, respectively. For example, 25Q032 was projected to enroll 1,014 students in 2021-2022, but only has the capacity to house 598 students, an excess of 416 students. In 2021-2022, the new school will enroll these 416 students that would have been enrolled in 25Q032, and 25Q032 will enroll 598 students in 2021-2022. The new school can house 84 more students (500 capacity minus 416 students from 25Q032), so the new school will enroll 84 students from the second more crowded school, 25Q184. 
    - Projected enrollment with new school: 
        - Modify the percent change in enrollment for the 2021-2022 school year because the 2020-2021 enrollment has been impacted by the new school. Update the percent change in rollment for the 2021-2022 school year to be the average percent change in enrollment from the 2019-2020 and 2020-2021 school years.

             - Projected enrollment in 2022-2023:

                 - 611.765866 students
